## Installing git

- on mac os : `$ git`

> if you have git, you will see the git help message
> if you don't have git , it will bring up a prompt to install Xcode Command Line Tools. Accept it. when the install process is finished, try again, you should have git available.

- on windows go to https://git-scm.com/downloads/win , choose standalone installer, 64 bits. during the installation process, click "next" and accept every proposition.

## Configuring git

- first thing to do after installing is to set your user name and email address. it's important because every Git commit use this information : the name of the developer must be added to every commit.
```
$ git config --global user.name "John Doe"
$ git config --global user.email johndoe@example.com
```

> if you've made a mistake, you can launch the command again to modify the value


- then configure vs code as default editor
> ```git config --global core.editor "code --wait"```
> please that this will work only if the `code` shortcut for vscode is active. if it is not, follow the steps [here](https://stackoverflow.com/questions/29955500/code-is-not-working-in-on-the-command-line-for-visual-studio-code-on-os-x-ma)

- when this is done, check your settings with :
```
git config --list
```

## The simplest way to do your first commit

This is what you will want to do after the first successful work session on the files of your project ( a notebook , some python code, a website ...).

You must type this shell commands with your terminal IN YOUR PROJECT DIRECTORY ( what you just did by following the precedent instructions )

- ```$ git add .```
- ```$ git commit -m "significant commit message"```


`git add .` will start to track all the content of your project folder, INCLUDING SUBFOLDERS, and put in the staging area all the modifications you've made.








## The simplest way to do the following commits



 Whenever you add a new feature that's worth committing, commit. You added a working method? Commit. You fixed a typo? Commit. You fixed a file's wrong indentation? Commit. There's nothing wrong committing small changes, as soon as the commit is relevant.

- if there is no new files or folders since the last "git add ." , you can simply do :  

>- ```$ git commit -am "significant commit message"```

>this will commit ALL the modifications made on the files already tracked, thanks to the "-a" switch of `git commit`. if you've added new files, they will not be part of the commit, even with `git commit -a`



- if there is new files to track , you must first explicitely ask git to start to track them using

>- ```$ git add .``` this will once again scan your project folder and subfolders for new files to track
>- ```$ git commit -m "significant commit message"``` ( you don't need the `-a` this time, because you've just asked git to stage all modifications.)






## Verifying things

- use ```$ git log``` to display the list of your commits.

## Creating ssh keys , to make publication to github easy



- checking if you already have an ssh key :

>```$ ls -la ~/.ssh```

> this will list all the files in the .ssh folder of your home directory. if you see id_ed25519.pub, you already have one. if not, do the following :


- ```$ ssh-keygen -t ed25519 -C "youremail@yourdomain.org"```

> then press enter 3 times , to accept the proposed file location, and set an empty passphrase for your key. this will generate your pair of ssh key, in the .ssh folder of your home directory

- ```$ cat ~/.ssh/id_ed25519.pub```

> the cat program is used to display the content of a file. here we display the content of the public key you've just created. cut it, and paste it here : https://github.com/settings/ssh/new
> this will add your public key to your github account.

- test your ssh connection with ```$ ssh -T git@github.com```

> you should see the word "successfully" somewhere in the output



## Creating a public github repository, and connecting our local repository to it.

On your github profile page, click on "repositories", then on the green button "new". you should see something like this :

You can choose a name for your project, or click on the proposed generated name in green. DO NOT CHANGE THE OTHER OPTIONS and click on "create a repository" below.

You should then see something like this :

image.png

- copy the ssh address given par github on the above page NOT THE HTTPS ADRESS. for my repository, the address is `git@github.com:robderon/studious-enigma.git`






- now we can connect our local repository to our remote github repository using git remote. ( don't forget that your terminal or git bash must be in your project directory for this to work) ( and of course, change the address with the address given by github for your repo)
```
$ git remote add origin git@github.com:robderon/studious-enigma.git
```
now, your git repository is connected to the github repository you've just created.

<br/>

- let's check our remote connections
```
$ git remote -v
```
the command should answer with two lines, meaning that we can push and fetch (we will explain this later) to and from this remote repo


- we can now **push** our code and show it to the world. let the magic begiins !
```
git push -u origin main
```
this will push our commited codebase, the branch named "main" to our online repository, named "origin".

- SOMETIMES THIS WILL NOT WORK, with "error: failed to push some refs". This is because your "branch" is not named `main`, but `master`. you then have two option, either respecting the new standard and renaming your branch "main" by doing

>```$ git branch -M main```

> either push using your existing branch name:

>```$ git push -u origin master```


- now reload your github repository page, your should see your files.

## How to ask git to ignore some files ?

When you use `git add .` to start tracking files, it will sometimes add files you don't want to include in your commits. for instance the .DS_Store on mac os, password files etc..

We must then specifically ask git to ignore these files. To do this, you must create a .gitignore file.

```$ code .gitignore```

this should open vscode, in the text file, write the following line

`.DS_Store`

and save the file. the .DS_Store file will now be ignored everytime. This will not work if the file you want to ignore have already been tracked. If you want to ignore a file that is already checked in, you must untrack the file before you add a rule to ignore it. From your terminal, untrack the file.

```git rm --cached FILENAME```
